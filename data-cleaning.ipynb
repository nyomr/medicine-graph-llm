```python
data = pd.read_csv("Medicine_Details.csv")
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
dfc = df
```

```python
dfc
```

```python
dfc['Composition'] = dfc['Composition'].str.replace(' + ', ', ', regex=False)
```

```python
dfc['Composition'] = dfc['Composition'].str.replace(r'\s*\(.*?\)', '', regex=True)
```

```python
dfc
```

```python
dfx = dfc
```

```python
print(dfx['Side_effects'].iloc[0])
```

```python
dfx['Side_effects'] = dfx['Side_effects'].astype(str)
```

```python
dfx['Side_effects'] = dfx['Side_effects'].str.replace(r'(?<=[a-z]) (?=[A-Z])', ', ', regex=True)
```

```python
print(dfx['Side_effects'].iloc[0])
```

```python
dfx.columns = dfx.columns.str.replace(' %', '', regex=False)
```

```python
dfx
```

```python
dfx.info()
```

```python
dfx.isnull().sum()
```

```python
dfn = dfx
```

```python
dfn.columns = dfn.columns.str.replace(' ', '')
```

```python
dfn.columns = dfn.columns.str.replace('_', '')
```

```python
dfn
```

```python
dfn.to_csv('medicine_cleaned.csv', index=False)
print("File CSV berhasil disimpan!")
```
